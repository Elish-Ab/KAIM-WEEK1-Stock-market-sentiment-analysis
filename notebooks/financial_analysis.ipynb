```python
# Set the correct directory path where the CSV files are located
data_dir = '/home/elisha-a/week1/yfinance_data'

# List of tickers corresponding to the datasets
tickers = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA', 'TSLA']

# Initialize an empty dictionary to store the data
stock_data = {}
```

```python
# Define the indicators to apply
indicators = [
    calculate_rsi,
    calculate_macd,
    calculate_bollinger_bands,
    calculate_ema,
    calculate_sma
]
```

```python
# Process each stock
for ticker in tickers:
    print(f"Processing data for {ticker}...")
    
    # Load the stock data
    file_path = os.path.join(data_dir, f"{ticker}_historical_data.csv")
    stock_data[ticker] = pd.read_csv(file_path, index_col='Date', parse_dates=True)
    
    # Debugging: Check loaded data columns
    print(f"Columns in {ticker} data: {stock_data[ticker].columns}")
    
    # Apply indicators
    for indicator in indicators:
        stock_data[ticker] = indicator(stock_data[ticker])
    
    # Save the processed data
    output_path = os.path.join(data_dir, f"{ticker}_with_indicators.csv")
    stock_data[ticker].to_csv(output_path)
    print(f"Processed data saved for {ticker} at {output_path}")
    
    # Visualizations
    print(f"Visualizing data for {ticker}...")
    plot_stock_data(stock_data[ticker], ticker)
    plot_macd(stock_data[ticker], ticker)
    plot_rsi(stock_data[ticker], ticker)
    plot_bollinger_bands(stock_data[ticker], ticker)
    plot_ema(stock_data[ticker], ticker)
```
